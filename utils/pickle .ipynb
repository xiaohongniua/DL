```python
import pickle
```

```python

class Person():
    def __init__(self, name, age):
        self.name = name
        self.age = age
 
    def show(self):
        return self.name + '_' + str(self.age)
```

```python
a = Person('tian', 20)
b=a.show()
```

```python
type(b)
```

```python
a
```

```python
c = Person('babydragon' ,5)
```

```python
with open('test.pkl', 'wb') as f:
    pickle.dump(a, f)
    pickle.dump(b, f)
    pickle.dump(c, f)
```

```python
with open('test2.pkl', 'w') as f:#这里一定要加上b,因为文件默认的write不能是bytes类型的，可以自己尝试着删去看报错
    pickle.dump(a, f)
    pickle.dump(b, f)
    pickle.dump(c, f)
```

```python
pickle.dump(a, 'test2_pkl')#file 参数是一个打开了的对象
```

```python
f=open('test.pkl', 'rb')
```

# 总结：
   ## 1.两个函数：pickle.dump(obj,file)，pickle.load(file);其中file参数都是打开了的文件对象对象，且是二进制模式
   ## 2.对象也是可以存取的
   ## 3. 多个对象可以同时存取，取出的时候依照存储的时候的顺序迭代取出
